```python
import matplotlib.pyplot as plt
import numpy as np

def function(x):
    return pow(x, 3) - 2 * pow(x, 2) + 4 * x - 8 + np.sin(x)

def deri(x):
    return 3 * pow(x, 2) - 4 * x + 4 + np.cos(x)

tol = 0.0001
max_iter = 1000
i = -1
diff = 100
roots = np.array([2])
x = roots[i + 1]

while (i < max_iter and abs(diff) > tol):
    i = i + 1
    x = roots[i]
    diff = function(x) / deri(x)
    x = x - diff
    roots = np.append(roots, x, axis = None)

plt.plot(roots)
plt.xlabel("Iterations")
plt.ylabel("Difference between successive roots")
plt.title("Plot of Convergence")
```

```python
x1 = np.random.normal(0, 1, 2000)
x2 = np.random.normal(1, 2, 2000)

x1mean = np.mean(x1)
print(x1mean)
x2mean = np.mean(x2)
print(x2mean)

x1median = np.median(x1)
print(x1median)
x2median = np.median(x2)
print(x2median)

x1var = np.var(x1)
print(x1var)
x2var = np.var(x2)
print(x2var)

x1std = np.std(x1)
print(x1std)
x2std = np.std(x2)
print(x2std)

bins = np.linspace(-10, 10, 100)

x1hist, _ = np.histogram(x1, bins, density = True)
x2hist, _ = np.histogram(x2, bins, density = True)

x1hist = np.where(x1hist == 0, 1e-10, x1hist) # replace 0 with small value
x2hist = np.where(x2hist == 0, 1e-10, x2hist)

plt.hist(x1, bins, alpha = 0.5)
plt.hist(x2, bins, alpha = 0.5)
plt.title("Histogram")

print("KL Divergence = ", np.sum(np.multiply(x1hist, np.log(np.divide(x1hist, x2hist)))))
```
